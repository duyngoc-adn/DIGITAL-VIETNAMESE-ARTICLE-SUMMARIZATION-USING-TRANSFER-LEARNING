```python
import requests
from bs4 import BeautifulSoup
import os
```

```python

def CrawData_TuoiTre(url):
    news = requests.get(url)
    soup = BeautifulSoup(news.content, "html.parser")
        
        # get title
    try: 
        title = soup.find("h1", class_="article-title").text.strip()
    except:
        print(os.error("Không tìm thấy"))
                    
        # get abstract
    try:
        abstract = soup.find("h2", class_="sapo").text
        if abstract.startswith("TTO - ") == True:
            abstract = abstract.replace("TTO - ", "", 1)
    except:
        print(os.error("Không tìm thấy"))
        
        # get body
    body = soup.find("div", id="main-detail-body")
    content = ""
    for i in body.findChildren("p", recursive=False):
        content += str(i.text.strip() + " ")
            
        # append data
    data = ({
        "title": title,
            "abstract": abstract.strip(),
            "content": content.strip(),
        })

    return data
```

# Hàm thu thập data từ báo Tuoitre.vn

```python
def crawNewsData_TuoiTre(baseUrl, url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    titles = soup.findAll('h3', class_='title-news')
    links = [link.find('a').attrs["href"] for link in titles]

    data = []
    for link in links:
        news = requests.get(baseUrl + link)
        soup = BeautifulSoup(news.content, "html.parser")
        
        # get title
        try: 
            title = soup.find("h1", class_="article-title").text.strip()
        except:
            continue
            
        # get abstract
        try:
            abstract = soup.find("h2", class_="sapo").text
            if abstract.startswith("TTO - ") == True:
                abstract = abstract.replace("TTO - ", "", 1)
        except:
            continue
        
        # get body
        body = soup.find("div", id="main-detail-body")
        content = ""
        for i in body.findChildren("p", recursive=False):
            content += str(i.text.strip() + " ")
        if content == "" :
            continue
            
        # append data
        data.append({
            "title": title,
            "abstract": abstract.strip(),
            "content": content.strip(),
        })

    return data
```
